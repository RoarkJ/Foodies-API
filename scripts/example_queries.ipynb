```python
import requests
import json
from pprint import pprint
import pandas as pd
```

## Filter data by Name and City

```python
name = "mcdonalds"
city = "arvada"
```

```python
base_url = "http://127.0.0.1:5000/api/v1.0/search/name_city"
response = requests.get(f"{base_url}/{name}/{city}")
```

```python
data = response.json()
# print all returned businesses
pprint(data)
```

```python
names = []
ratings = []
prices = []
address1 = []
citys = []
states = []
zip_codes = []

for restaurant in data:
    name = restaurant["restaurant_name"]
    rating = restaurant["rating"]
    price = restaurant["price"]
    address_1 = restaurant["location"]["address1"]
    city = restaurant["location"]["city"]
    state = restaurant["location"]["state"]
    zip_code = restaurant["location"]["zip_code"]
    hours = restaurant["hours"]
    
    names.append(name)
    ratings.append(rating)
    prices.append(price)
    address1.append(address_1)
    citys.append(city)
    states.append(state)
    zip_codes.append(zip_code)
    
data = {
    "restaurant_name":names,
    "rating":ratings,
    "prices": prices,
    "street_address":address1,
    "city": citys,
    "state":states,
    "zip_code":zip_codes
}

pd.DataFrame(data)
```
