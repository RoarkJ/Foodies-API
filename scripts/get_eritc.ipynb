```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas
import pymongo
import time
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client.restaurant_db
```

```python
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# pulling data for colorado restaurants only
browser.visit("https://everyrestaurantinthecity.com/colorado-restaurants")
urls = []

# retrive URL's for each city in Colorado
for page in range(5):
    time.sleep(2)
    soup = BeautifulSoup(browser.html, "html.parser")
    table = soup.find("table", style="width:100%")
    cities = table.find_all("p")
    # url = [city.find("a")["href"]for city in cities if city.find("a")["href"]]
    for city in cities:
        if city.find("a")["href"]:
            urls.append(city.find("a")["href"])
    try:
        browser.find_by_text(" Next ").click()
    except:
        print("Completed URL retrieval")
```

```python
# used to retrieve restaurant data
def get_data(restaurants):
    for restaurant in restaurants:
        try:
            # retrieving information
            restaurant_name = restaurant.find("td").find("strong").text
            restaurant_phone = restaurant.find("div", class_="icon_search").text
            # breaking down the address into individual pieces
            address_list = restaurant.find("td").text.split("-")[1].strip()
            address_list = address_list.split(",")
            # retrieving address information
            restaurant_state = address_list[0].strip()
            restaurant_city = address_list[1].strip()
            street_address = address_list[2].strip()
            restaurant_zipcode = address_list[3].strip()
            
            # create object to insert into database
            post = {
                "name": restaurant_name,
                "phone": restaurant_phone,
                "street_address": street_address,
                "city": restaurant_city,
                "state": restaurant_state,
                "zipcode": restaurant_zipcode
            }

            # insert object into database
            db.restaurant_atomic.insert_one(post)
        except:
            print("Record Not Found")
```

```python
for url in urls:
    # visit webpage which lists restaurants in a specific city
    browser.visit(url)
    time.sleep(3)
    
    # if only one page of results
    if not browser.find_by_text(" Next "):
        # soupify that page's html
        soup = BeautifulSoup(browser.html, "html.parser")
        # extract information
        restaurants = soup.find_all("table", class_="box")
        # retrieve data
        get_data(restaurants)
    
    # if more than one page of results
    elif browser.find_by_text(" Next "):
        # while there is a next page of results
        while browser.find_by_text(" Next "):
            # soupify page's html
            soup = BeautifulSoup(browser.html, "html.parser")
            # extract information
            restaurants = soup.find_all("table", class_="box")
            # retrieve data
            get_data(restaurants)
            # move to next page
            browser.find_by_text(" Next ").click()
            # when bot has reached last page
            if not browser.find_by_text(" Next "):
                # soupify html of last page
                soup = BeautifulSoup(browser.html, "html.parser")
                # extract information on last page
                restaurants = soup.find_all("table", class_="box")
                # retrieve data
                get_data(restaurants)
    time.sleep(3)
```
